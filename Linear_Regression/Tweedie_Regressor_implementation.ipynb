```python
col_list_to_be_included =[]
col_list_to_be_included.extend(col_list[1:-2])
col_list_to_be_included.append(col_list[-1])
col_list_to_be_included
```

```python
train_df1 = train_df[col_list_to_be_included]
target = train_df['Attrition_rate']
```

```python
target = np.log(1+target)
```

```python
train_df1_columns = list(train_df1.columns)
test_df_columns = list(test_df.columns)
train_df1_columns,test_df_columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
train_df1 = pd.DataFrame(ss.fit_transform(train_df1), columns=train_df1_columns)
train_df1.head()
```

```python
test_df.iloc[:,1:] = pd.DataFrame(ss.transform(test_df.iloc[:,1:]),columns=test_df_columns[1:])
```

```python
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, valid_x , train_y, valid_y = train_test_split(train_df1,target, test_size=0.10,random_state=43)
```

```python
from sklearn.linear_model import TweedieRegressor
```

```python
tw_reg = TweedieRegressor(power=2)
```

```python
tw_reg.fit(train_x,(train_y+1))
```

```python

import eli5
```

```python
from eli5.sklearn import PermutationImportance
```

```python
model = PermutationImportance(tw_reg)
```

```python
model.fit(train_x,(train_y+1))
```

```python
eli5.show_weights(model, feature_names = train_x.columns.tolist())
```

```python
train_x.columns
```

```python
selected_features = list(train_x.columns)
```

```python
tw_reg = TweedieRegressor(power=2)
```

```python
tw_reg.fit(train_x[selected_features],(train_y+1))
```

```python
pred_values = tw_reg.predict(valid_x[selected_features])
```

```python
pred_values = pred_values-1
```

```python
pred_values = np.expm1(pred_values)
```

```python
pred_values[:5]
```

```python
pred_values = np.round(pred_values,4)
```

```python
pred_values[:5]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
100*max(0,(1-np.sqrt(mean_squared_error(valid_y,pred_values))))
```
